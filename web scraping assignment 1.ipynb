IMDB's top 100 movies

```python
movie = requests.get('https://www.imdb.com/chart/top')
```

```python
movie
```

```python
soup = BeautifulSoup(movie.text)
```

```python
soup
```

```python
import pandas as pd
movies = []
for i in soup.find_all('td',class_='titleColumn'):
    movies.append(i.text.split(','))
movies = movies[:100]
movies
```

```python
dates = []
for i in soup.find_all('span',class_='secondaryInfo'):
    dates.append(i.text)
dates = dates[:100]
dates
```

```python
rating = []
for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(i.text)
rating = rating[:100]
rating
```

```python
print(len(movies),len(dates),len(rating))
```

```python
scrapemovie = pd.DataFrame({'Movie':movies,'Year':dates,'IMDB Rating':rating})
scrapemovie
```

IMDB Top Rated 100 Indian Movies

```python
indmovie = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
indmovie
```

```python
indsoup = BeautifulSoup(indmovie.text)
```

```python
indsoup
```

```python
indmovies = []
for i in indsoup.find_all('td',class_='titleColumn'):
    indmovies.append(i.text.replace(',',' '))
indmovies = indmovies[:100]
indmovies
```

```python
year = []
for i in indsoup.find_all('span',class_='secondaryInfo'):
    year.append(i.text.split(','))
year = year[:100]
year
```

```python
ratings = []
for i in indsoup.find_all('td',class_='ratingColumn imdbRating'):
    ratings.append(i.text.split(','))
ratings = ratings[:100]
ratings
```

```python
print(len(indmovies),len(year),len(ratings))
```

```python
scrape_ind_movie = pd.DataFrame({'Top 100 Indian Movie':indmovies,'Year of Release':year,'IMDB Rating':ratings})
scrape_ind_movie
```

Scraping Cricket Rankings

```python
cricrank = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
cricrank
```

```python
cric = BeautifulSoup(cricrank.text)
```

```python
cric
```

```python
top_odi = []
for j in cric.find_all('tr',class_='rankings-block__banner'):
    top_odi.append(j.text)
for i in cric.find_all('tr',class_='table-body'):
    top_odi.append(i.text)
top_odi = top_odi[:10]
top_odi
```

```python
top_10_odi = pd.DataFrame({'TOP 10 ODI SIDES AND RATINGS':top_odi})
top_10_odi
```
