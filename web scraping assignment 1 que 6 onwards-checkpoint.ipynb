```python
from bs4 import BeautifulSoup
import requests
```

To find mobiles under 20000 from Amazon.in

```python
mob = requests.get('https://www.amazon.in/s?k=mobiles+under+20000rs%2B&crid=8OAVRNGBUC0E&sprefix=mobiles+under+20000%2Caps%2C1055&ref=nb_sb_ss_ts-doa-p_1_19')
```

```python
mob
```

```python
soup = BeautifulSoup(mob.content)
```

```python
soup
```

```python
import pandas as pd
mobile = []
for i in soup.find_all('span',class_='a-size-medium a-color-base a-text-normal'):
    mobile.append(i.text.split('|')[0])
mobile
```

```python
price = []
for i in soup.find_all('span',class_='a-price-whole'):
    price.append(i.text.split('|')[0])
price
```

```python
images = []
for i in soup.find_all('img',class_='s-image'):
    images.append(i['src'])
images = images[:16]
images
```

```python
rating = []
for i in soup.find_all('div',class_='a-row a-size-small'):
    rating.append(i.text.split(',')[0])
rating = rating[:16]
rating
```

```python
print(len(mobile),len(price),len(images),len(rating))
```

```python
mobile_20000 = pd.DataFrame({'Mobile':mobile,'Price':price,'Image':images,'Avg Rating':rating})
mobile_20000
```

to find restaurents from https://www.dineout.co.in/delhi-restaurants/buffet-special

```python
dineout = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
dineout
```

```python
soup2 = BeautifulSoup(dineout.content)
```

```python
soup2
```

```python
restname = []
for i in soup2.find_all('div', class_='restnt-info cursor'):
    restname.append(i.text.split(',')[0])
restname
```

```python
cuisine = []
for i in soup2.find_all('div',class_='detail-info'):
    cuisine.append(i.text.split('|')[1])
cuisine
```

```python
location = []
for i in soup2.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text.split('|')[0])
location
```

```python
rating = []
for i in soup2.find_all('div', class_='restnt-rating rating-3'):
    rating.append(i.text)
for j in soup2.find_all('div',class_='restnt-rating rating-4'):
    rating.append(j.text)
rating
```

```python
image = []
for i in soup2.find_all('img',class_='no-img'):
    image.append(i['data-src'])
image
```

```python
print(len(restname),len(cuisine),len(location),len(rating),len(image))
```

```python
dineouts = pd.DataFrame({'Restaurent Name':restname,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Images':image})
dineouts
```

Weather Info from https://en.tutiempo.net/delhi.html?data=last-24-hours

```python
weather = requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')
```

```python
weather
```

```python
soup3 = BeautifulSoup(weather.content)
```

```python
soup3
```

```python
wc = []
for i in soup3.find_all('span',class_='thhip ico i0530 u303'):
    wc.append(i.text)
wc = wc[:24]
wc
```

```python
temp = []
for i in soup3.find_all('td',class_='t Temp'):
    temp.append(i.text)
temp = temp[:24]
temp
```

```python
wind = []
for i in soup3.find_all('td',class_='wind'):
    wind.append(i.text)
wind = wind[:24]
wind
```

```python
hum = []
for i in soup3.find_all('td',class_='hr'):
    hum.append(i.text)
hum = hum[:24]
hum
```

```python
pressure = []
for i in soup3.find_all('td',class_='prob'):
    pressure.append(i.text)
pressure = pressure[:24]
pressure
```

```python
print(len(temp),len(wc),len(wind),len(hum),len(pressure))
```

```python
weather_24 = pd.DataFrame({'Temperature':temp,'Weather Condition':wc,'Wind Speed':wind,'Humidity':hum,'Air Pressure':pressure})
```

```python
weather_24
```
